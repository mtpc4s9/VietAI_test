```python
import numpy as np
```

```python
def calculate_c(target_p):
    # C is the constant that makes the average probability equal to target_p
    return target_p * target_p / 2

def get_prd_probability(n, C):
    # n is the number of failures
    # Returns probability for next attempt
    return min(1.0, n * C)
```

```python
# Target probability = 15%
target_p = 0.15
C = calculate_c(target_p)

# Calculate initial probability (n=1)
initial_p = get_prd_probability(1, C)
```

```python
# Print probabilities for first 10 attempts
print("Probability progression:")
for n in range(1, 11):
    p = get_prd_probability(n, C)
    print(f"After {n} misses: {p*100:.2f}%")
```

```python
# Calculate expected number of attempts until success
def expected_attempts():
    attempts = 0
    total_prob = 0
    n = 1
    while total_prob < 0.95:  # 95% confidence
        p = get_prd_probability(n, C)
        total_prob += p * (1 - total_prob)
        attempts += 1
        n += 1
    return attempts
```

```python
exp_attempts = expected_attempts()
print(f"\nExpected attempts until 95% confidence: {exp_attempts}")
```
